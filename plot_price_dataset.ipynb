```python
import pandas as pd 
df=pd.read_excel(uploaded['plot price dataset (1).xlsx'])
x=df['plot size']
y=df['plot Price']
len(x)
```

```python
class MylR:
  def fit(self,x,y,epoch,learningrate):
    m=0
    c=0
    length=len(x)
    for j in range(epoch):
      for i in range(length):
        model_output=m*x[i]+c
        error=model_output-y[i]
        gm=-2*error*x[i]/length
        gc=-2*error/length
        m=m+(learningrate*gm)
        c=c+(learningrate*gc)
    return m,c
  def predict(self,m,c,x):
    output=[]
    for i in range(len(x)):
      o=m*x[i]+c
      output.append(o)
    return output
model=MylR()
m,c=model.fit(x,y,100,0.001)
O=model.predict(m,c,x)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,O,c='red')
```
